```python
tickers = get_day_gainers()['Symbol'].tolist()
dic={}
for ticker in tickers :
    try :
        dic[ticker] = get_data(ticker, start_date='01/01/2017', end_date='14/07/2021')
    except :
        pass
    
df = reduce(lambda x,y : x.append(y), list(dic.values()))
```

```python
drawdown={}
for ticker in tickers:
    try :
        current_price = df[df['ticker']==ticker]['adjclose'].iloc[-1]
        max_price = df[df['ticker']==ticker]['adjclose'].max()
        drawdown[ticker] = (current_price-max_price)/max_price
    except :
        pass
```

```python
df_drawdown = pd.DataFrame(data=list(drawdown.values()), index=list(drawdown.keys()), columns=['drawdown'])
```

```python
min_.values[0] = df_drawdown.min()
df_drawdown[df_drawdown['drawdown']==min_.values[0]]
df_drawdown.sort_values(by='drawdown')
```
